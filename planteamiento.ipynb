```python
board = [5,3]

'''
En la definición del problema, el tablero se define mediante una tupla con el tamaño de cada dimensión [vertical, horizontal].

Es importante destacar que no es el valor de array sino la dimensión del mismo
'''

# check board restrictions
board_error = False
board_error_text = ''

# chek dimension
if len(board) == 2: 
#     print('Board dimension is ok. ')
    pass
else:
    board_error_text += 'Incorrect board dimension. '
    board_error = True

# chek size
for n in range(2):
    if board[n] >= 1 and board[n] <= 10:
#         print('Size of dimension ' + str(n+1) + ' is correct. ')
        pass
    else:
        board_error_text += 'Size of dimension ' + str(n+1) + ' is out of bound. '
        board_error = True

# evaluate if error        
if board_error:
    print(board_error_text + 'Correct board array and try again.')

```

```python
snake = [[4, 2], [4, 1], [4, 0]]



'''
A diferencia del tablero, los segmentos de la serpiente se definen por las coordenadas de array (indice inicial 0)
'''

# chek snake restrictions
snake_error = False
snake_error_text = ''

# chek length
if len(snake) >= 3 and len(snake) <= 7: 
#     print('Snake length is ok.')
    pass
else:
    snake_error_text += 'Incorrect snake length.'
    snake_error = True

# chek body dimmensions
for n in range(len(snake)):
    if len(snake[n]) == 2: 
#         print('Snake body segment ' + str(n+1) + ' dimension is ok. ')
        pass
    else:
        snake_error_text += 'Incorrect snake body segment ' + str(n+1) + ' dimension. '
        snake_error = True

# check body continuity
for n in range(1,len(snake)):
    d0n0 = snake[n-1][0]
    d0n1 = snake[n][0]
    d1n0 = snake[n-1][1]
    d1n1 = snake[n][1]
    if (d0n0 == d0n1 and abs(d1n0-d1n1) == 1) or (d1n0 == d1n1 and abs(d0n0-d0n1) == 1):
#         print('Snake body junction ' + str(n-1) + '-' + str(n) + ' has a correct continuity. ')
        pass
    else:
        snake_error_text += 'Snake body junction ' + str(n-1) + '-' + str(n) + ' has a discontinuity. '
        snake_error = True

def body_outside_boundaries(board, snake):
    for n in range(len(snake)):
        for m in range(2):
            if snake[n][m] < 0 or snake[n][m] >= board[m]:
                return True
    return False

if body_outside_boundaries(board, snake):
    snake_error_text += 'Snake body segment is out of board dimension. '
    snake_error = True
else:
#     print('Snake body segment ' + str(n+1) + ' is correctly inside board dimension ' + str(m+1) + '. ')
    pass

# check body clash
def body_clash(snake):
    set_snake = set()
    for n in range(len(snake)):
        if tuple(snake[n]) in set_snake:
            return True
        else:
            set_snake.add(tuple(snake[n]))         
    return False

if body_clash(snake):
    snake_error_text += 'Snake segment is clashing. '
    snake_error = True
        
# evaluate if error 
if snake_error:
    print(snake_error_text + 'Correct snake body arrays and try again.')
```

```python
depth = 5

# chek depth restrictions
depth_error = False
depth_error_text = ''

# check size
if depth >= 1 and depth <= 20:
#     print('Path depth is correct. ')
    pass
else:
    depth_error_text += 'Path depth is out of length. '
    depth_error = True
    
# evaluate if error 
if depth_error:
    print(depth_error_text + 'Correct the path depth and try again.')
```

```python
def move_snake(snake, direction):
    new_head = [[x + y for x, y in zip(snake[0], conver_direction(direction))]]
    new_body = snake[0:len(snake)-1]
    new_snake = new_head + new_body
    return new_snake

def conver_direction(command):
    if command == 'L':
        move = [0,-1]
    if command == 'U':
        move = [-1,0]
    if command == 'R':
        move = [0, 1]
    if command == 'D':
        move = [1, 0]
    return move

def viable_solution(board,snake,direction):
    n_snake = move_snake(snake, direction)
    if body_outside_boundaries(board, n_snake):
        return False
    if body_clash(n_snake):
        return False
    return True

def modulo_10_p_7(n,p):
    ''' Inicialmente se hizo una función para determinar el módulo 10^9+7, pero la velocidad de procesamiento no es
    óptima y he optado por cambiarlo para que se pueda decidir en cada caso el valor del módulo P'''
    M = 10**p+7
    return n % M
```

```python
def numberOfAvailableDifferentPaths(board, snake, depth, solution = [], step = 0, j = 0):
    MOVES = ['L', 'U', 'R', 'D']
    solution = ([''] * depth) if not solution else solution
    n_snake = snake.copy()
    for i in MOVES:
        snake = n_snake.copy()
        solution[step] = i
        if viable_solution(board, snake, i):
            snake = move_snake(snake, i)
            if step == depth -1:
                j = j + 1
                print('[ play',j,']:', solution)
            else:
                j = numberOfAvailableDifferentPaths(board, snake, depth, solution, step + 1, j)
        else:
#             print(i, '|||', snake)
            pass
        solution[step] = ''
        if j > modulo_10_p_7(j, 3): return j
    return j
  
```
